```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
f = h5py.File(
    "/home/b08209033/Projects/julia/Dynamical_Core/Experiment/output/Datasets_2025_1028_171312.jld2",
    "r",
)
for item in f.keys():
    print(item)
```

```python
lat = np.rad2deg(f["latitude"][:])
lon = np.rad2deg(f["longitude"][:])
num_daily_data = np.arange(0, 360, 1)
vor = f["vor_hourly_mean"][:].squeeze()
h = f["h_hourly_mean"][:].squeeze()

lat_mask = np.where((lat > 0) & (lat < 40))[0]
lat_mask = slice(lat_mask[0], lat_mask[-1] + 1)

lon_mask = np.where((lon > 0) & (lon < 360))[0]
lon_mask = slice(lon_mask[0], lon_mask[-1] + 1)
```

```python
time = 359
plt.contour(
    lon[lon_mask],
    lat[lat_mask],
    h[time, lat_mask, lon_mask],
    levels=10,
    colors="k",
)
plt.contourf(
    lon[lon_mask],
    lat[lat_mask],
    vor[time, lat_mask, lon_mask],
    levels=10,
    cmap="jet",
)
```
