```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 40, endpoint=False)
y = np.cos(1*x) + np.cos(2*x) + np.cos(3*x)
y_padded = np.concatenate([np.zeros(180,),y, np.zeros(180,)])
```

```python
x_test = np.linspace(0, 2*np.pi, 400, endpoint=False)
y_test = np.cos(10*x_test) + np.cos(20*x_test) + np.cos(30*x_test)
y_test[40:] = 0.0
y_test = np.roll(y_test, 180)
```

```python
plt.figure()
plt.plot(y)
plt.figure()
plt.bar(np.fft.fftshift(np.fft.fftfreq(40,1/40)),np.fft.fftshift(np.fft.fft(y)))
plt.xlim(-5,5)
plt.figure()
plt.plot(y_padded)
plt.figure()
plt.bar(np.fft.fftshift(np.fft.fftfreq(400,1/400)),np.fft.fftshift(np.fft.fft(y_padded)))
plt.xlim(-50,50)
plt.figure()
plt.plot(y_test)
plt.figure()
plt.bar(np.fft.fftshift(np.fft.fftfreq(400,1/400)),np.fft.fftshift(np.fft.fft(y_test)))
plt.xlim(-50,50)
```
